## Simple Experiment 

Suppose that $f(x) = GP(\mu, k(x,x'))$, for simplicity we set $\mu = 0$, but the covariance $k(x,x')$ is unknown. Now suppose that we chose the following functional form for the kernel:

$$k(x,x') = \sigma^2_f \exp\bigg( \frac{- (x - x')^2}{2l^2}\bigg) $$

It is clear from this that our two hyper parameters are $\sigma^2_f$ and $l$. Intuitivley, the covariance function determines the 'shape' of the function that we are estimating. It does this by dictating how different the distance between outputs is, given a pair of inputs. For a simple explanation see: [Gaussian Processes for Dummies - Katherine Bailey](http://katbailey.github.io/post/gaussian-processes-for-dummies/)

```python
### CALCULATING EUCLIDEAN (KM) DISTANCE ###

#we need to use a package to calculate distance 
#it doesn't make a lot of sense to calculate distance with long/lats
#package makes use of 'haversine' distance

#relevant package
import geopy.distance

#creating a function for ease of use

def euclidean_distance(x, y):
    '''Calculates Euclidean Distance between two tuples x and y'''
    # we can't just use Euclidean formula because of Earth's curvature
    z = geopy.distance.distance(x, y).km
    return(z)

def reverse_tuple(x):
    '''function that reverses the order of a tuple'''
    y = (x[1], x[0])
    return(y)

```

```python
#### TOY EXAMPLE ####
#[1] Generate a df around a geographic region that is feasible
#[2] Calculate Euclidean distance between points
#[3] Construct Gaussian Process for fixed hyperparameters
#[4] Plot

def SA2_to_SA1(SA2code):
    '''returns the SA1 codes that aggregate to an SA2 region'''
    df = meshblocks[meshblocks.SA2_NAME_2016 == SA2code]
    #we must use unique since the dataframe contains codes relating to meshblocks which are smaller
    y = list(df.SA1_7DIGITCODE_2016.unique())
    return(y)

def GSYD(code):
    '''returns the SA1 codes that aggregate to a GCCSA Code'''
    df = meshblocks[meshblocks.GCCSA_NAME_2016 == code]
    #we must use unique since the dataframe contains codes relating to meshblocks which are smaller
    y = list(df.SA1_7DIGITCODE_2016.unique())
    return(y)
    
G_list = ['Greater Sydney']

inner_west_list = ['Marrickville', 
                   'Petersham - Stanmore', 
                   'Erskinevile - Alexandria', 
                   'Glebe - Forrest Lodge', 
                   'Newtown - Camperdown - Darlington',
                   'Redfern - Chippendale']

#returns a list of lists
sa1 = [SA2_to_SA1(x) for x in inner_west_list]
G = [GSYD(x) for x in G_list]

#returns a 'flat' list, useful when len(inner_west_list) > 1 
sa1 = [str(item) for sublist in sa1 for item in sublist]
G = [str(item) for sublist in G for item in sublist]

#for each item in sa1 we want to obtain latittudes/longtitudes and centroids
#subsetting upon the large geo_df we constructed

inner_west = df[df['SA1_7DIGITCODE_2016'].isin(G)]

#isolating centroids
training_X = inner_west['centroids']

#obtaining coordinates
training_X = [x.coords[0] for x in training_X]

#reversing to properly format
training_X = [reverse_tuple(x) for x in training_X]

#splitting into lattitude and longtitudes
training_X1 = [x[0] for x in training_X]
training_X2 = [x[1] for x in training_X]

#creating into an appropriate array
training_X = np.array(training_X).reshape(len(training_X),2)

#obtaining labels
y = np.array([np.float(x) for x in inner_west['Median_rent_weekly'] ])
y = y.reshape(len(y),1)
```

```python
#modelling the GP with a fixed parameter

k = gpflow.kernels.Matern52(2, lengthscales=0.3)
m = gpflow.models.GPR(training_X, y, kern=k)
m.likelihood.variance = 1
```

```python
##obtaining training data
# - we want 100 training data points
# - 0.5 long/lat from newtown centroid
# - covariance matrix will therefore be 53 x 100 matrix

#-33.8970, 151.1793 original we are taking +- 0.5
#if we use np.arange the floating point round off will mess things up
x1 = np.linspace(np.mean(training_X1) + 0.5, np.mean(training_X1) - 0.5, num = 5) #x
x2 = np.linspace(np.mean(training_X2) - 0.5 ,  np.mean(training_X2) + 0.5,num = 5) #y
xx1, xx2 = np.meshgrid(x1,x2)
Xplot = np.vstack((xx1.flatten(),xx2.flatten())).T

```

```python
#fitting the GP for fixed parameters
mean, var = m.predict_f(Xplot)
```

```python
## CREDIT: https://www.kaggle.com/ankasor/driving-distance-using-open-street-maps-data


def street_distance(area_graph, start, end):
    #Find nodes closest to the specified Coordinates
    node_start = ox.utils.get_nearest_node(area_graph, start)
    node_stop = ox.utils.get_nearest_node(area_graph, end)
    
    #Calculate the shortest network distance between the nodes via the edges "length" attribute
    distance = nx.shortest_path_length(area_graph, node_start, node_stop, weight = "length")
    
    return(distance)
```

```python
#making a new Kernel -- we copied from the existing RBF kernel source code
#in order to make a new Kernel, we need to obtain street network distance 

import tensorflow as tf


class SE_Euclid(gpflow.kernels.Kernel):
    """
    Squared Exponential Kernel utilising Euclidean Distance
    """

#Note: Parameters need to be tensors

    def __init__(self, variance, lengthscale):
        super().__init__(input_dim= 2)
        self.variance = gpflow.Param(1.0, transform=gpflow.transforms.positive)
        self.lengthscale = gpflow.Param(1.0, transform=gpflow.transforms.positive)

    def square_dist(self, X, X2):
        X = X / self.lengthscale
        Xs = tf.reduce_sum(tf.square(X), axis=1)

        if X2 is None:
            dist = -2 * tf.matmul(X, X, transpose_b=True)
            dist += tf.reshape(Xs, (-1, 1)) + tf.reshape(Xs, (1, -1))
            return (dist)

        X2 = X2 / self.lengthscale
        X2s = tf.reduce_sum(tf.square(X2), axis=1)
        dist = -2 * tf.matmul(X, X2, transpose_b=True)
        dist += tf.reshape(Xs, (-1, 1)) + tf.reshape(X2s, (1, -1))
        return (dist)

    def euclid_dist(self, X, X2):
        r2 = self.square_dist(X, X2)
        return (tf.sqrt(r2 + 1e-12))

    def K(self, X, X2=None, presliced=False):
        if not presliced:
            X, X2 = self._slice(X, X2)
            return (self.variance * tf.exp(- self.square_dist(X, X2) * (1 / self.lengthscale ** 2)))
    def Kdiag(self, X, presliced=False):
        return(tf.fill(tf.stack([tf.shape(X)[0]]), tf.squeeze(self.variance)))
```
